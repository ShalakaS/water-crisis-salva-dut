```python
GPS = df[df['GPS '] != 0]
GPS
```

```python
GPS['GPS '].str.len().loc[400:430].reset_index()
```

```python
decimals = GPS[~GPS['GPS '].str.contains("[']")]

split_cols = decimals['GPS '].str.split('E:', expand=True)
```

```python
split_cols = split_cols.add_prefix('lat')
```

```python
split_cols.columns = split_cols.columns.str.replace("lat0", "lat").str.replace("lat1", "long")
```

```python
decimals_1 = pd.concat([decimals, split_cols], axis=1)
decimals_1.lat = decimals_1.lat.str.replace(" ", "").str.replace("N:", "").str.replace('["]', "")
decimals_1.long = decimals_1.long.str.replace(" ", "").str.replace('["]', "")
decimals_1 = decimals_1.drop(columns=['GPS '])
decimals_1
```

```python
minutes = GPS[GPS['GPS '].str.contains("[']")]
minutes
```

```python
split_cols2 = minutes['GPS '].str.split('E:', expand=True)
split_cols2 = split_cols2.add_prefix('lat')
```

```python
split_cols2.columns = split_cols2.columns.str.replace("lat0", "lat").str.replace("lat1", "long")
```

```python
split_cols2['long'] = split_cols2['long'].apply(lambda x: 'E:' + str(x))
```

```python
minutes_1 = pd.concat([minutes, split_cols2], axis=1) 
minutes_1 = minutes_1.drop(columns =['GPS '])
minutes_1
```

```python
minutes_1.head()
```

```python
split_lats = minutes_1.lat.str.split(' ', expand=True)
split_lats = split_lats.drop(columns=[0, 4, 5, 6, 7])
split_lats = split_lats.add_prefix('lat')
split_lats
```

```python
split_lats.lat1 = split_lats.lat1.str.replace("[']", "")
```

```python
split_lats.lat1 = split_lats.lat1.str.replace(".", "").str.replace(";", "").str.replace(" ", "")
```

```python
split_lats.lat2 = split_lats.lat2.str.replace("[']", "")
```

```python
split_lats.lat3 = split_lats.lat3.str.replace('["]', '').str.replace('238:', '238')
```

```python
split_lats.lat1 = pd.to_numeric(split_lats['lat1'], errors='coerce').fillna(0).astype(float)
```

```python
split_lats.lat2 = pd.to_numeric(split_lats['lat2'], errors='coerce').fillna(0).astype(float)
```

```python
split_lats.lat3 = pd.to_numeric(split_lats['lat3'], errors='coerce').fillna(0).astype(float)
```

```python
split_lats['lat_cons'] = split_lats.lat1 + (split_lats.lat2/60) + (split_lats.lat3/3600)
```

```python
split_lats = split_lats.drop(columns=["lat1", "lat2", "lat3"])
split_lats
```

```python
split_longs = minutes_1.long.str.split(' ', expand=True)
split_longs = split_longs.drop(columns=[0, 4, 5, 6])
split_longs = split_longs.add_prefix('long')
split_longs
```

```python
split_longs.long1 = split_longs.long1.str.replace("[']", "").str.replace(" ", "")
```

```python
split_longs.long1 = split_longs.long1.astype(float)
```

```python
split_longs.long2 = split_longs.long2.str.replace("[']", "").str.replace(" ", "").str.replace('["]', '')
```

```python
split_longs.long3 = split_longs.long3.str.replace('["]', '').str.replace('.', '').str.replace("[']", "")
```

```python
split_longs.long2 = pd.to_numeric(split_longs['long2'], errors='coerce').fillna(0).astype(float)
split_longs.long3 = pd.to_numeric(split_longs['long3'], errors='coerce').fillna(0).astype(float)
```

```python
split_longs['long_cons'] = split_longs.long1 + (split_longs.long2/60) + (split_longs.long3/3600)
split_longs
```

```python
split_longs = split_longs.drop(columns=["long1", "long2", "long3"])
split_longs
```

```python
minutes_2 = pd.concat([minutes_1, split_lats, split_longs], axis=1) 
minutes_2 = minutes_2.drop(columns=["lat", "long", "coords"])
minutes_2.columns = minutes_2.columns.str.replace("_cons", "")
minutes_2
```
