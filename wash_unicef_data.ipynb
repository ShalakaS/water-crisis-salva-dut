```python
import pandas as pd
```

```python
df = pd.read_csv('washdata.csv')
df
```

```python
rural = df[df['Residence Type'] == 'rural'].reset_index(drop=True)
rural
```

```python
rural.groupby(['Year', 'Service Level']).Coverage.mean().reset_index()
```

```python
rural_long = pd.melt(rural, id_vars=['Service Level', 'Year'], value_vars=['Coverage'])
rural_long.columns = rural_long.columns.str.replace(" ", "")
rural_long
```

```python
import numpy as np
```

```python
rural_pivot = pd.pivot_table(rural, values='Coverage', index=['Year'],
                       columns=['Service Level'], aggfunc=np.sum)
rural_pivot
```

```python
rural_pivot = rural_pivot.reset_index()
rural_pivot
```
